```python
import numpy as np

# Example beer servings data
beer_servings = [200, 150, 300, 180, 400, 220, 250, 1000, 280, 160, 170, 1500, 230]

# Using IQR method
Q1 = np.percentile(beer_servings, 25)
Q3 = np.percentile(beer_servings, 75)
IQR = Q3 - Q1
lower_threshold = Q1 - 1.5 * IQR
upper_threshold = Q3 + 1.5 * IQR
iqr_outliers = [x for x in beer_servings if x < lower_threshold or x > upper_threshold]

# Using Z-score method
mean = np.mean(beer_servings)
std_dev = np.std(beer_servings)
z_scores = [(x - mean) / std_dev for x in beer_servings]
z_score_threshold = 3
z_score_outliers = [beer_servings[i] for i, z in enumerate(z_scores) if abs(z) > z_score_threshold]

print("Outliers using IQR method:", iqr_outliers)
print("Outliers using Z-score method:", z_score_outliers)
```
